```python
%matplotlib inline
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import json
import datetime
import collections
matplotlib.style.use('ggplot')
plt.rcParams['figure.figsize'] = [12.0, 8.0]
```

```python
dates = pd.date_range('2017-03-12','2017-03-26')
```

```python
dir_name = './step/'
time_in_step = []
step_time = {}
for date in dates:
    fname = dir_name+str(date.year)+'-'+str(date.month).zfill(2)+'-'+str(date.day).zfill(2)
    with open(fname) as f:
        date_data = json.loads(f.read())
        time_in_step.append(int(date_data['activities-steps'][0]['value']))
        for intraday in date_data['activities-steps-intraday']['dataset']:
            if intraday['time'] not in step_time.keys():
                step_time[intraday['time']] = int(intraday['value'])
            else:
                step_time[intraday['time']] += int(intraday['value'])
df = pd.DataFrame(time_in_step,index=dates)
df.columns=['step']
df.plot()
plt.ylabel('steps in a Day')
```

```python
df.describe().transpose()
```

```python
step_time_od  = collections.OrderedDict(sorted(step_time.items()))
```

```python
df = pd.DataFrame(list(step_time_od.values()),index=list(step_time_od.keys()))
df.columns=['step']
df.plot()
plt.ylabel('steps per 15min')
```

```python
dir_name = './step/'
days = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 
        5: 'Saturday', 6: 'Sunday'}
step_days = {}
step_date = {} 
i=1
for date in dates:
    fname = dir_name+str(date.year)+'-'+str(date.month).zfill(2)+'-'+str(date.day).zfill(2)
    with open(fname) as f:
        date_data = json.loads(f.read())
        activities_step = date_data['activities-steps'][0]['value']
        if date.dayofweek not in step_days.keys():
            step_days[date.dayofweek] = int(activities_step)
        else:
            step_days[date.dayofweek] = ( step_days[date.dayofweek] *(i-1) +int(activities_step) ) / i
    i=i+1
```

```python
step_days_od  = collections.OrderedDict(sorted(step_days.items()))
```

```python
df = pd.DataFrame(list(step_days_od.values()),index=list(step_days_od.keys()))
```

```python
days = ['월','화','수','목','금','토','일']
fig, ax = plt.subplots()
fig.suptitle('요일별 이용 횟수 비교')
ind = np.arange(len(days))
width = 0.35
rects1 = ax.bar(ind,df.values,width, color='r')
ax.set_xticks(ind)
ax.set_xticklabels(days)
plt.show()
```
